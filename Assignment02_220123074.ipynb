```python
def retlist(N_of_rand_num):
    m=128
    b=55
    a=17 #values chosen according to knuth law to get a good approaximation of U(0,1)
    x0=24
    A=np.zeros(N_of_rand_num)
    for i in range(N_of_rand_num):
        ui=x0/m
        A[i]=1-(1-ui)**(1/3)
        x0=(a*x0+b)%m
    return A
'''
This function will return list of size N of distribution corresponding to first type
'''



```

```python
def CDF(list , x0): #first argument is the list of values second is the argument to CDF 
    n=len(list)
    cnt=0
    for i in range(n):
        if(list[i]<=x0):
            cnt+=1
    return cnt/n
#This function is to calculate CDF plot is done later 
```

```python
def returnlist(Num_of_rand_num):
    m=2**15
    a=65
    b=729 #again according to knuth law
    x0=1
    A=np.zeros(Num_of_rand_num)
    for i in range(Num_of_rand_num):
       ui=x0/m
       if(ui==0):
          A[i]=0
       elif(ui<=(1-1/math.e)):
           A[i]=math.log(1/(1-ui))
       else:
           A[i]=0.5*(1-math.log(1-ui))
       x0=(a*x0+b)%m
    return A 
```

```python
b1=returnlist(10)
b2=returnlist(100)
b3=returnlist(1000)
b4=returnlist(10000)
b5=returnlist(100000)
```
