```python
#TASK 01 part 01
a=6
b=0
m=11
# formula to generate random numbers x(i+1)=(ax(i)+b)%m
list=[]
for x0 in range(0,11):
    temp=[x0]
    input=-1
    while(input!=x0):
        input=(a* temp[-1]+b)%m
        if(input!=x0):
            temp.append(input)
    list.append(temp)
for i in range(11):
    print("if we choose x0=" + str(i)+" we get the sequence ")
    print(list[i])
    print("with " + str(len(list[i]))+" numbers before it repeats again")
```
